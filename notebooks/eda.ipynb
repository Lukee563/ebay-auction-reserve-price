## Distributional Analysis of Reserve Price (openbid)
In this dataset, 'openbid' is the seller's opening bid, which functions as the reserve price (the minimum price the seller is willing to accept). No bids below it will win the auction.

It's worth noting though that eBay actually has two distinct concepts that often get conflated:
- Opening bid: the starting price publicly visible to bidders, which is what openbid captures here
- Secret reserve price: a hidden minimum that eBay also allows sellers to set separately, which this dataset doesn't include

For this analysis, 'openbid' will serve as the reserve price proxy.

```python
print(auction['openbid'].describe())
```

The distribution of reserve price is heavily right skewed, with a median of $\$4.99$ and a maximum value of $\$5,000$. The mean is $\$52.25$, pulled up by a high-end outliers. The 25th percentile at $\$1.00$ tells us a large share of sellers are essentially setting no meaningful floor at all.

In our regressions we will log reserve price to account for the heavily skewed support of our variable. 

## Kernel Density Plot (Raw vs Log)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Raw
sns.kdeplot(auction['openbid'], ax=axes[0], fill=True)
axes[0].set_title('Reserve Price Distribution (Raw)')
axes[0].set_xlabel('Opening Bid (\$)')

# Log-transformed
sns.kdeplot(np.log1p(auction['openbid']), ax=axes[1], fill=True)
axes[1].set_title('Reserve Price Distribution (Log)')
axes[1].set_xlabel('log(Opening Bid + 1)')

plt.tight_layout()
plt.show()
```
